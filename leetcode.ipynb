```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        count = 1
        majority = nums[0]

        for i in range(1,n):
            if nums[i]==majority:
                count += 1
            else : 
                count -=1
                if count ==0:
                    majority = nums[i]
                    count = 1

        count = 0
        for i in range(n):
            if nums[i] == majority:
                count += 1
        if count > n//2:
            return majority
        else:
            return -1
        
```

```python
# findMissingAndRepeatedValues

from typing import List

class Solution(object):
    def findMissingAndRepeatedValues(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: List[int]
        """
        n = len(grid)
        N = n*n
        actualSum = 0
        actualSquareSum = 0
        #  compute actual sum
        for i in range(n):
            for j in range(n):
                num = grid[i][j]
                actualSum += num
                actualSquareSum += num*num

        # compute expected sum
        expectedSum = (N*(N+1))//2
        expectedSquareSum = (N*(N+1)*(2*N+1))//6

        # calcute difference

        sumDifference = actualSum-expectedSum

        squareSumDifference = actualSquareSum -expectedSquareSum

        # use derived equation to find the missing and repeated values

        sum_ab = squareSumDifference // sumDifference

        repeated = (sum_ab + sumDifference)//2
        missing = (sum_ab- sumDifference)//2

        return [repeated,missing]
    
    # Run the function with a sample grid
if __name__ == "__main__":
   # Example: 1 is repeated, 3 is missing
   grid = [
       [1, 2],
       [1, 4]
   ]


   sol = Solution()
   result = sol.findMissingAndRepeatedValues(grid)


   print("Repeated:", result[0])
   print("Missing:", result[1])

        
```

```python
# findDuplicate
class Solution(object):
    def findDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        slow, fast = nums[0],nums[0]

        # move slow by one and fast by 2 step
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]

            if slow == fast:
                break

         # reset slow to the start and move both pointer
        slow = nums[0]
        while slow != fast:
            fast = nums[fast]
            slow = nums[slow]

        return slow
```

```python
# largestRectangleArea
class Solution(object):
    def largestRectangleArea(self, heights):
        """
        :type heights: List[int]
        :rtype: int
        """
        n = len(heights)
        index_stack=[]
        max_area=float('-inf') # negative infinity

        # iterate through the bar
        for i in range(n):
            while index_stack and heights[i] < heights[index_stack[-1]]:
                top_index=index_stack.pop()
                height=heights[top_index]

                # calculate width
                width = i if not index_stack else(i-index_stack[-1]-1)
                area = height*width

                # update maximum area
                max_area=max(max_area,area)

            # push current index to the stack
            index_stack.append(i)

        # handle remaining bar in the stack
        while index_stack:
            top_index = index_stack.pop()
            height = heights[top_index]

            # calculate width for the remaining elements
            width = n if not index_stack else(n-index_stack[-1]-1)
            area = height*width

            # update max area
            max_area  = max(area,max_area)

        return max_area
```

```python
#  4 Sum
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        n = len(nums)
        result = []
        # loop for first number
        for first in range(n-3):
            if first > 0 and nums[first] == nums[first-1]:
                continue

            # loop for second number
            for second in range(first+1,n-2):
                if second > first+1 and nums[second]==nums[second]:
                    continue

                remainingTarget = target-nums[first]-nums[second]
                left,right = second+1,n-1


                # two pointer approach
                while left < right :
                    sum_ = nums[left]+nums[right]

                    if sum_ < remainingTarget:
                        left+=1

                    elif sum_ > remainingTarget:
                        right -=1

                    else:
                        result.append([nums[first],nums[second],nums[left],nums[right]])

                        # skip duplicate
                        prevLeft ,prevRight = nums[left],nums[right]

                        while left<right and nums[left]==prevLeft:
                            left+=1
                        while left<right and nums[right]==prevRight:
                            right-=1
        return result


```

```python
# removeOccurrences
class Solution(object):
    def removeOccurrences(self, s, part):
        """
        :type s: str
        :type part: str
        :rtype: str
        """
        while part in s:
            idx = s.find(part)
            s = s[:idx] + s[idx + len(part):]

        return s
```

```python
# String compression
class Solution(object):
    def compress(self, chars):
        """
        :type chars: List[str]
        :rtype: int
        """
        chars.append('~')
        n= len(chars)
        i=0
        count =1
        for j in range(1,n):
            if chars[j]==chars[j-1]:
                count +=1
            else:
                chars[i] = chars[j-1]
                i+= 1
                if count>=2:
                    for c in str(count):
                        chars[i]=c
                        i+=1
                    count = 1
        return i
```

```python
# checkInclusion
class Solution(object):
    def checkInclusion(self, s1, s2):
        """
        :type s1: str
        :type s2: str
        :rtype: bool
        """
        n1 = len(s1)
        n2 = len(s2)

        if n1 > n2:
            return False

        count1 = [0]*26
        count2 = [0]*26

        for i in range(n1):
            count1[ord(s1[i])-ord('a')]+=1
            count2[ord(s2[i])-ord('a')]+=1

        if count1 == count2:
            return True

        j = 0
        for i in range(n1,n2):
            count2[ord(s2[i])-ord('a')]+=1
            count2[ord(s2[j])-ord('a')]-=1
            if count1 == count2:
                return True
            j+=1
        return  False

```

```python
# groupAnagrams
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        ans = []
        n = len(strs)

        from collections import defaultdict

        m = defaultdict(list)
        for i in range(n):
            str_ = strs[i]
            sorted_str = ' '.join(sorted(strs[i]))
            m[sorted_str].append(str_)

        for group in m.values():
            ans.append(group)
        return ans
```

```python
# minWindow

class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        target = [0]* 256 #frequency of character in t

        for ch in t : 
            target[ord(ch)]+=1

        n = len(s)
        i,j =0,0
        count = 0
        required = len(t)
        start = 0
        min_len = float('inf')

        while j<n:
            if target [ord(s[j])]>0:
                count+=1
            
            target [ord(s[j])]-=1

            while count == required:
                if min_len >j-i+1:
                    min_len = j-i+1
                    start = i
                
                target[ord(s[i])]+=1
                if target[ord(s[i])]>0:
                    count -= 1
                i+=1
            j+=1


        return "" if min_len == float('inf') else s[start:start + min_len]

```

```python
# Search in sorted Matrix

class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        m = len(matrix)
        n = len(matrix[0])
        r = 0
        c = n-1

        while c>=0 and r<m:
            if target == matrix[r][c]:
                return True
            elif target<matrix[r][c]:
                c-=1
            else:
                r+=1
        return False
```

```python
class Solution(object):
    def minimizedMaximum(self, n, quantities):
        """
        :type n: int
        :type quantities: List[int]
        :rtype: int
        """
        def canDistribute(maxProducts):
            storesNeeded = 0
            for q in quantities:
                storesNeeded += (q + maxProducts-1)//maxProducts

            return storesNeeded<=n
        low = 1
        high = max(quantities)
        answer = high

        while low<=high:
            mid=(low+high)//2

            if canDistribute(mid):
                answer=mid
                high=mid-1
            else:
                low=mid+1
        return answer

        
            
```

```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""

        reference = strs[0]

        for index, expected_char in enumerate(reference):
            for word in strs:
                if index == len(word) or word[index] != expected_char:
                    return reference[:index]

        return reference
```

```python
class Solution(object):
    def singleNonDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left = 0 
        right = len(nums)-1

        while left< right:
            mid = left + (right - left)//2

            # make the mid even so it point to the expected first item of a pair

            if mid%2 == 1:
                mid -= 1
            if nums[mid] == nums[mid+1]:
                left = mid+2

            else:
                right = mid

        return nums[left]
```
